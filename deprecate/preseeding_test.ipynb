```python
from slam.basis import CircuitTemplate
basis = CircuitTemplate(preseed=True, use_polytopes=True)
basis.clear_all_save_data()
```

```python
from slam.cost_function import BasicCost
objective = BasicCost()
```

```python
from slam.optimizer import TemplateOptimizer
optimizer = TemplateOptimizer(basis=basis, objective=objective, use_callback=True)
len(basis.data_dict)
```

```python
from slam.sampler import HaarSample as HaarSample #Haar3Sample
# for i in range(10):
#     sampler = HaarSample(n_samples=1, seed=i)
#     ret = optimizer.approximate_from_distribution(sampler=sampler)
sampler = HaarSample(n_samples=1, seed=0)
ret = optimizer.approximate_from_distribution(sampler=sampler)
```

```python
from slam.utils.visualize import optimizer_training_plot
optimizer_training_plot(*ret);
```
