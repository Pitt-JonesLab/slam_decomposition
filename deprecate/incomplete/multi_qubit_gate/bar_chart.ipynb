```python
import numpy as np
# import pandas as pd
# from pandas import Series, DataFrame
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,4))
ax = fig.subplots(1,1)
x = np.array([7,8,9,10])
data1 = [0,0, 41,48]
data2 = [0,0,35,49]
data3 = [2,42,40,50]
data4 = [49,49,50,50]
width =0.1
ax.bar(x, data1, width=width, label = 'CCX')
ax.bar(x+width, data2, width=width, label = 'iCCX')
ax.bar(x+2*width, data3, width=width, label = 'V-iSwap')
ax.bar(x+3*width, data4, width=width, label = 'Î”-iSwap')
ax.set_xlabel('Variational Circuit Depth')
ax.set_ylabel("Decomposition Success Rate (N=50)")
ax.legend()
ax.set_xticks([7,8,9,10])
plt.rcParams.update({'font.size': 14})
ax.legend(bbox_to_anchor=(1,1), ncol=1)
plt.tight_layout()
fig.savefig("barchart.svg", format='svg', facecolor='None')
fig.show()
```
